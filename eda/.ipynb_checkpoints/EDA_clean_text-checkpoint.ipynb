```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import string
import re
import nltk

from tqdm import trange
from nltk import tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.probability import FreqDist
from collections import Counter
from sklearn.feature_extraction.text import CountVectorizer
```

```python
import warnings
warnings.filterwarnings('ignore')
nltk.download('omw-1.4', quiet=True)
sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = (17,7)
plt.rcParams['font.size'] = 18
```

```python
# load data
data = pd.read_excel('../data/captions_autocaptions_clean.xlsx')
data.head()
```

```python
data['Objectify'].value_counts()
```

```python
def ratings(rating):
    if rating == 0:
        return "No"
    else:
        return "Yes"
data['Rating'] = data['Objectify'].apply(ratings)
```

### EDA on text

```python
lenght_captions = len(data['clean_caption'][0])
print(f'Length of processed captions: {lenght_captions}')
lenght_autocaptions = len(data['clean_autocaption'][0])
print(f'Length of processed autocaptions: {lenght_autocaptions}')
lenght_text = len(data['All_text'][0])
print(f'Length of combined: {lenght_text}')
```

```python
data['Length_caption'] = data['clean_caption'].str.len()
data['Length_autocaption'] = data['clean_autocaption'].str.len()
data['Length_all'] = data['All_text'].str.len()
data.head(10)
```

```python
# word count
def word_count(review):
    review_list = review.split()
    return len(review_list)
```

```python
data['All_text'] = data['All_text'].astype(str)
```

```python
data['Word_count'] = data['All_text'].apply(word_count)
data.head()
```

```python
data['mean_word_length'] = data['All_text'].map(lambda rev: np.mean([len(word) for word in rev.split()]))
data.head()
```

```python
#mean sentence length
np.mean([len(sent) for sent in tokenize.sent_tokenize(data['All_text'][0])])
```

```python
data['mean_sent_length'] = data['All_text'].map(lambda rev: np.mean([len(sent) for sent in tokenize.sent_tokenize(rev)]))
data.head()
```

Check for NaNs

```python
data = data.dropna()
```

```python
def visualize(col):
    
    print()
    plt.subplot(1,2,1)
    sns.boxplot(data = data, y = col, hue='Rating')
    plt.ylabel(col, labelpad=12.5)
    
    plt.subplot(1,2,2)
    sns.kdeplot(data = data,x = col, hue='Rating')
    plt.legend(data['Rating'].unique())
    plt.xlabel('')
    plt.ylabel('')
    
    plt.show()
```

```python
data.columns
```

```python
data
```

```python
features = ['Length_caption', 'Length_autocaption', 'Length_all',
       'Word_count', 'mean_word_length', 'mean_sent_length']
for feature in features:
    print(feature)
    visualize(feature)
```

## Term Frequency Analysis

```python
df = data.drop(features, axis=1)
df = df.reset_index(drop = True)
```

```python
def corpus(text):
    text_list = text.split()
    return text_list

df['Text_lists'] = df['All_text'].apply(corpus)
df.head()
```

```python
df['Text_lists']
```

```python
corpus = []
for i in trange(df.shape[0], colour='green', smoothing=0.8):
    corpus += df['Text_lists'][i]
len(corpus)
```

```python
mostCommon = Counter(corpus).most_common(10)
mostCommon
```

```python
words = []
freq = []
for word, count in mostCommon:
    words.append(word)
    freq.append(count)
```

```python
sns.barplot(x=freq, y=words)
plt.title('Top 10 Most Frequently Occuring Words')
plt.show()
```

### N-grams

2 grams

```python
cv = CountVectorizer(ngram_range=(2,2))
bigrams = cv.fit_transform(df['All_text'])
```

```python
count_values = bigrams.toarray().sum(axis=0)
ngram_freq = pd.DataFrame(sorted([(count_values[i], k) for k, i in cv.vocabulary_.items()], reverse = True))
ngram_freq.columns = ["frequency", "ngram"]
```

```python
sns.barplot(x=ngram_freq['frequency'][:10], y=ngram_freq['ngram'][:10])
plt.title('Top 10 Most Frequently Occuring Bigrams')
plt.show()
```

3 grams

```python
cv1 = CountVectorizer(ngram_range=(3,3))
trigrams = cv1.fit_transform(df['All_text'])
count_values = trigrams.toarray().sum(axis=0)
ngram_freq = pd.DataFrame(sorted([(count_values[i], k) for k, i in cv1.vocabulary_.items()], reverse = True))
ngram_freq.columns = ["frequency", "ngram"]
```

```python
sns.barplot(x=ngram_freq['frequency'][:10], y=ngram_freq['ngram'][:10])
plt.title('Top 10 Most Frequently Occuring Trigrams')
plt.show()
```
