```python
import os

def find_txt_files(directory):
    txt_files = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.txt'):
                txt_files.append(os.path.join(root, file))
    return txt_files

# Example usage
directory_path = '../1_download_data/data/raw'
txt_files = find_txt_files(directory_path)
txt_files
```

```python
import os


def organize_jpg_files(directory):
    jpg_files_dict = {}
    for root, dirs, files in os.walk(directory):
        # Extracting subfolder names
        for dir in dirs:
            subfolder_path = os.path.join(root, dir)
            jpg_files = [file for file in os.listdir(subfolder_path) if file.endswith('.jpg')]
            if jpg_files:
                jpg_files_dict[dir] = jpg_files
    return jpg_files_dict


directory_path = '../1_download_data/data'
jpg_files_dict = organize_jpg_files(directory_path)
```

### Turn to table and merge

```python
def create_dataframe_from_dict(data_dict):
    data = []
    for key, values in data_dict.items():
        for value in values:
            data.append([key, value])
    df = pd.DataFrame(data, columns=['Subfolder', 'FileName'])
    return df


df_long_codes = create_dataframe_from_dict(jpg_files_dict)

df_long_codes
```

```python
df_long_codes.columns = ['Short', 'Long']
df_long_codes['Long'] = df_long_codes['Long'].apply(lambda x: x.split('.jpg')[0])
df_long_codes.head()
```
