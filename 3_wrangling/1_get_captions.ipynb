Get txt files

```python
def get_txt_files(folder_path, users):
    txt_files = []
    for user in users:
        user_folder = os.path.join(folder_path, user)
        if os.path.exists(user_folder):
            for root, dirs, files in os.walk(user_folder):
                for file in files:
                    if file.endswith('.txt'):
                        txt_files.append(os.path.join(root, file))
        else:
            print(f"Folder not found: {user_folder}")
    return txt_files

# Example usage
users = ['hm', 'jacquemus', 'sheinofficial', 'chanelofficial']
folder_path = '../1_download_data/'
txt_files = get_txt_files(folder_path, users)
txt_files[0:10]
```

```python
path = ''
dict_captions = {}
for filename in txt_files:
    file_path = os.path.join(path, filename)  # It's better to use os.path.join

    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            contents = file.read()
        dict_captions[filename] = contents  # Use filename as the key

    except UnicodeDecodeError:
        try:
            with open(file_path, 'r', encoding='latin-1') as file:
                contents = file.read()
            dict_captions[filename] = contents  # Use filename as the key

        except UnicodeDecodeError:
            # Handle the case where both encodings fail
            print(f"Could not decode the file {filename} with utf-8 or latin-1 encoding.")
            dict_captions[filename] = None  # You might want to assign None or a placeholder
```

```python
# Create a DataFrame
df = pd.DataFrame(list(dict_captions.items()), columns=['posts', 'caption'])

#df_captions = df.merge(df_posts, how = 'left', on='Long')
#df_captions
#df['posts'][0].split('.txt')[0].split()
df.head()
```
