```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_face = pd.read_excel('../0_data/dataframe_face_detection.xlsx')
df_face.head()
```

```python
df_skin = pd.read_excel('../0_data/dataframe_skin_pct.xlsx')
df_skin.head()
```

```python
df_features = df_face.merge(df_skin, how = 'inner', on ='Long')
df_features.isna().any()
```

```python
df_features['Long'] = df_features['Long'].apply(lambda x: x.split('.')[0])
```

```python
df_text = pd.read_excel('../0_data/pictures_captions_autocaptions_v2.xlsx')
df_final = df_features.merge(df_text, how = 'right', on='Long')
df_final.isna().any()
```

```python
df_final
```

```python
# Now, to find and display rows with NaN values
rows_with_nans = df_final[df_final.isna().any(axis=1)]
rows_with_nans
```

```python
df_final = df_final.dropna().reset_index(drop = True)
df_final
```

```python
df_final.to_excel('../0_data/data_model.xlsx', index = False)
```
