```python
import pandas as pd
import numpy as np
```

```python
df_autogen = pd.read_excel('../0_data/autogen_captions.xlsx')
df_autogen.head()
```

```python
df_autogen['Long'] = df_autogen['Long'].apply(lambda x:x.split('.')[0])
df_autogen
```

```python
df_codes = pd.read_excel('../0_data/dataframe_codigos.xlsx')
df_codes.head()
```

```python
df_codes = df_codes[['Long', 'Short']]
```

```python
print(len(df_codes))
print(len(df_autogen))
```

```python
# merge dataframes to get shortcodes
df_merge1 = df_autogen.merge(df_codes, how ='left', on = 'Long')
df_merge1.head()
```

```python
df_merge1.columns
```

```python
df_merge1.columns = ['Long', 'Auto_caption', 'Short']
```

## Colapse photos into one post

```python
joint_text = {}
for code in df_merge1['Short']:
    df_temp = df_merge1[df_merge1['Short'] == code]
    joined_string = df_temp['Auto_caption'].str.cat(sep=' ')
    joint_text[code] = joined_string
len(joint_text)
```

```python
# Convert to DataFrame
df_joint_text_auto = pd.DataFrame.from_dict(joint_text, orient='index').reset_index()

# Rename columns
df_joint_text_auto.columns = ['Short', 'Auto_caption']
df_joint_text_auto
```

Import regular captions and join with long captions

```python
df_captions = pd.read_excel('../0_data/dataframe_captions.xlsx')
df_captions.head()
```

```python
df_final = df_joint_text_auto.merge(df_captions, how = 'left', on = 'Short')
df_final
```

```python
df_final['All_text'] = df_final['caption'] + ' ' + df_final['Auto_caption']
df_final
```

```python
df_obj = pd.read_excel('../0_data/dataframe_codigos.xlsx')
df_obj = df_obj[['Short', 'Objectify (yes or no)']]
df_obj.columns = ['Short', 'Objectify']
df_obj.head()
```

```python
df_save = df_final.merge(df_obj, how='left', on='Short')
df_save.head()
```

```python
#drop duplicates
df_dropped_duplicates = df_save.drop_duplicates()
df_dropped_duplicates
```

```python
df_dropped_duplicates.groupby(['Objectify'])[['Objectify']].count()
```

```python
df_dropped_duplicates.to_excel('../0_data/captions_autocaptons.xlsx', index = False)
```
