```python
import pandas as pd
import numpy as np
```

Get processed data including autocaptions

```python
df = pd.read_excel('../0_data/bronze/5_objectification_data.xlsx')
df.head()
```

Get captions

```python
df_captions = pd.read_excel('../0_data/bronze/5_captions.xlsx')
df_captions.head()
```

```python
df_captions['posts'][0]
```

```python
df_captions['post_dir'] = df_captions['posts'].str.rsplit('\\', n=1).str[0]
df_captions
```

```python
print(len(df))
print(len(df_captions))
```

```python
# merge dataframes to get shortcodes
df_merge = df.merge(df_captions, how ='left', on = 'post_dir')
print(len(df_merge))
df_merge.head()
```

```python
df_merge['objectifies'].value_counts()
```

```python
sum(df_merge['objectifies'].isna())
```

```python
# drop nans because there is a data collection problem / also drop if not 0 or 1
df_clean = df_merge[df_merge['objectifies'].isin([0,1])]
print(len(df_merge))
print(len(df_clean))
print('Rows droped:', len(df_merge) - len(df_clean))
```

```python
df_clean.to_excel('../0_data/silver/6_data_silver_long.xlsx', index = False)
```

```python
# Check for duplicated rows
duplicates = df_clean.duplicated()

sum(duplicates)
```

```python
df_clean['post_dir'][0]
```

```python
df_clean
```

## Colapse photos into one post

```python
joint_text = {}
for code in df_clean['post_dir']:
    df_temp = df_clean[df_clean['post_dir'] == code]
    joined_string = df_temp['Auto_caption'].str.cat(sep=' ')
    joint_text[code] = joined_string
len(joint_text)
```

```python
# Convert to DataFrame
df_joint_text_auto = pd.DataFrame.from_dict(joint_text, orient='index').reset_index()

# Rename columns
df_joint_text_auto.columns = ['post_dir', 'Auto_caption']
#df_joint_text_auto = df_merge1.copy() # For Professor of Comunication
df_joint_text_auto
```

Import regular captions and join with long captions

```python
df_final_short = df_joint_text_auto.merge(df_captions, how = 'left', on = 'post_dir')
df_final_short
```

```python
df_final_short['All_text'] = df_final_short['caption'] + ' ' + df_final_short['Auto_caption']
df_final_short['All_text'][0]
```
