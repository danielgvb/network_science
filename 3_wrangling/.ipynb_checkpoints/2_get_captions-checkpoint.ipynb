```python
import pandas as pd
import numpy as np
import os
```

```python
df_posts = pd.read_excel('dataframe_codigos.xlsx')
df_posts.drop(columns = ['Unnamed: 0'], inplace = True)
```

```python
# Specify the directory you want to search
directory = 'downloaded_posts'

# List all files ending with .xz in the specified directory
txt_files = [file for file in os.listdir(directory) if file.endswith('.txt')]

# Print the list of files
txt_files
```

```python
path = 'downloaded_posts'
dict_captions = {}
for filename in txt_files:
    file_path = os.path.join(path, filename)  # It's better to use os.path.join

    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            contents = file.read()
        dict_captions[filename] = contents  # Use filename as the key

    except UnicodeDecodeError:
        try:
            with open(file_path, 'r', encoding='latin-1') as file:
                contents = file.read()
            dict_captions[filename] = contents  # Use filename as the key

        except UnicodeDecodeError:
            # Handle the case where both encodings fail
            print(f"Could not decode the file {filename} with utf-8 or latin-1 encoding.")
            dict_captions[filename] = None  # You might want to assign None or a placeholder
```

```python
dict_captions
```

```python
# Create a DataFrame
df = pd.DataFrame(list(dict_captions.items()), columns=['posts', 'caption'])
df['Long'] = df['posts'].apply(lambda x: x.split('.')[0])
df_captions = df.merge(df_posts, how = 'left', on='Long')
df_captions
```
