```python
import os

# Get the current working directory

# Specify the folder path you want to search for text files in
folder_path =  '../0_data/downloaded_posts'

# List all files in the folder
files = os.listdir(folder_path)

# Filter out only the text files (files with a .txt extension)
txt_files = [file for file in files if file.endswith('.txt')]
len(txt_files)
```

```python
import os
# Folder to save the posts
download_folder = 'downloaded_posts'
os.makedirs(download_folder, exist_ok=True)
shortcodes = []
ids = df_links['id']
dict_ids_codes = {}
for link, id in zip(links, ids):
    shortcode = link.split('/')[-2]
    dict_ids_codes[id] = shortcode    
```

```python
import os
import json
import lzma

# Specify the directory you want to search
directory = '../0_data/downloaded_posts'
codes_dict = {}
# List all files ending with .xz in the specified directory
xz_files = [file for file in os.listdir(directory) if file.endswith('.xz')]

for file in xz_files:
    # Open and decompress the .xz file
    file_short = file.split('.')[0]
    file_path = directory + '/' + file
    with lzma.open(file_path, 'rt') as file:
        data = json.load(file)
        
    codes_dict[file_short] = data['node']['shortcode']
```

### Turn to table and merge

```python
df_long_codes = pdf = pd.DataFrame(list(codes_dict.items()), columns=['Long', 'Short'])
df_long_codes.head()
```

```python
df_short_codes = pdf = pd.DataFrame(list(dict_ids_codes.items()), columns=['id', 'Short'])
df_short_codes.head()
```

```python
df_merge = df_short_codes.merge(df_long_codes, how='left', on = 'Short')
print(len(df_merge))
```
