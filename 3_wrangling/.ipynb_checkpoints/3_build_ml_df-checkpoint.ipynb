```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import data of faces
df_face = pd.read_excel('../0_data/bronze/3_data_faces.xlsx')
df_face.head()
```

```python
# import data of skin percentage
df_skin = pd.read_excel('../0_data/bronze/4_data_skin_pct.xlsx')
df_skin.head()
```

```python
# merge both features
df_features = df_face.merge(df_skin, how = 'inner', on ='Long')
df_features.isna().any()
```

```python
# rename column
df_features.columns = ['img_path', 'face', 'skin']
```

Import dataset of text and labels

```python
data = pd.read_excel('../0_data/silver/7_text_clean_large.xlsx')
data.drop(columns = ['Unnamed: 0'], inplace = True)
data.head()
```

```python
data.columns
```

```python
# keep only relevant data
df_text = data[['img_path', 'text_deep_clean', 'objectifies']]
df_text.head()
```

```python
df_text['objectifies'].value_counts()
```

```python
# merge datasets
df_final = df_features.merge(df_text, how = 'right', on='img_path')
df_final.isna().any()
```

```python
df_final
```

```python
# Now, to find and display rows with NaN values
rows_with_nans = df_final[df_final.isna().any(axis=1)]
rows_with_nans
```

```python
df_final = df_final.dropna().reset_index(drop = True)
df_final
```

```python
df_final.to_excel('../0_data/gold/8_data_model.xlsx', index = False)
```
