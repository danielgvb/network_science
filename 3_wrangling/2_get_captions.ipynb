Get txt files

```python
import os

def find_txt_files(directory):
    txt_files = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.txt'):
                txt_files.append(os.path.join(root, file))
    return txt_files

# Example usage
directory_path = '../1_download_data/data/raw'
txt_files = find_txt_files(directory_path)
txt_files
```

```python
path = ''
dict_captions = {}
for filename in txt_files:
    file_path = os.path.join(path, filename)  # It's better to use os.path.join

    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            contents = file.read()
        dict_captions[filename] = contents  # Use filename as the key

    except UnicodeDecodeError:
        try:
            with open(file_path, 'r', encoding='latin-1') as file:
                contents = file.read()
            dict_captions[filename] = contents  # Use filename as the key

        except UnicodeDecodeError:
            # Handle the case where both encodings fail
            print(f"Could not decode the file {filename} with utf-8 or latin-1 encoding.")
            dict_captions[filename] = None  # You might want to assign None or a placeholder
```

```python
# Create a DataFrame
df = pd.DataFrame(list(dict_captions.items()), columns=['posts', 'caption'])

#df_captions = df.merge(df_posts, how = 'left', on='Long')
#df_captions
#df['posts'][0].split('.txt')[0].split()
```

```python
def get_string_between_backslashes(input_string):
    # Splitting the string by backslashes
    parts = input_string.split('\\')
    # Return the string between the first two backslashes if it exists
    return parts[1] if len(parts) > 2 else ""

df['Short'] = df['posts'].apply(lambda x: get_string_between_backslashes(x))
df.head()
```
