```python
import pandas as pd
import numpy as np
```

```python
df_autogen = pd.read_excel('../0_data/autogenerated_clean.xlsx')
df_autogen.head()
```

```python
df_codes = pd.read_excel('../0_data/long_short_code_equivalence.xlsx')
df_codes.head()
```

```python
print(len(df_codes))
print(len(df_autogen))
```

```python
# merge dataframes to get shortcodes
df_merge1 = df_autogen.merge(df_codes, how ='left', on = 'Long')
df_merge1.head()
```

```python
df_merge1.columns
```

```python
df_merge1.columns = ['Long', 'Auto_caption', 'clean_text_auto', 'text_sup_clean_auto',
       'text_deep_clean_auto', 'text_deep_clean_pos_auto', 'hashtags_auto',
       'text_deep_clean_pos_str_auto', 'Short']
```

## Colapse photos into one post

```python
joint_text = {}
for code in df_merge1['Short']:
    df_temp = df_merge1[df_merge1['Short'] == code]
    joined_string = df_temp['text_deep_clean_pos_str_auto'].str.cat(sep=' ')
    joint_text[code] = joined_string
len(joint_text)
```

```python
# Convert to DataFrame
df_joint_text_auto = pd.DataFrame.from_dict(joint_text, orient='index').reset_index()

# Rename columns
df_joint_text_auto.columns = ['Short', 'text_deep_clean_pos_str_auto']
df_joint_text_auto
```

Import regular captions and join with long captions

```python
df_captions = pd.read_excel('../0_data/captions_clean.xlsx')
df_captions.drop(columns =['Unnamed: 0', 'Long'], inplace = True)
df_captions.head()
```

```python
df_merge2 = df_joint_text_auto.merge(df_captions, how = 'left', on = 'Short')
df_merge2
```

```python
df_merge2.columns
```

```python
# drop useless cols
df_final = df_merge2[['id','Short',
                      'Objectify (yes or no)', 
                      'text_deep_clean_pos_str', 'text_deep_clean_pos_str_auto']]
df_final.head()
```

```python
df_final['All_text'] = df_final['text_deep_clean_pos_str'] + ' ' + df_final['text_deep_clean_pos_str_auto']
df_final.head()
```

```python
df_final.columns = ['id','Short',
                      'Objectify', 
                      'clean_caption', 'clean_autocaption', 'All_text']
df_final
```

```python
df_final.groupby(['Objectify'])[['Objectify']].count()
```

```python
df_final.to_excel('../0_data/captions_autocaptons_clean.xlsx', index = False)
```
