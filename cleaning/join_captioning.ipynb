```python
import pandas as pd
import numpy as np
```

```python
df_autogen = pd.read_excel('../data/autogenerated_clean_v2.xlsx')
df_autogen.head()
```

```python
df_codes = pd.read_excel('../data/long_short_code_equivalence.xlsx')
df_codes.head()
```

```python
print(len(df_codes))
print(len(df_autogen))
```

```python
# merge dataframes to get shortcodes
df_merge1 = df_autogen.merge(df_codes, how ='left', on = 'Long')
df_merge1.head()
```

```python
df_merge1.columns
```

```python
df_merge1.columns = ['Long', 'Auto_caption', 'clean_text_auto', 'text_sup_clean_auto',
       'text_deep_clean_auto', 'text_deep_clean_pos_auto', 'hashtags_auto',
       'text_deep_clean_pos_str_auto', 'Short']
```

```python
df_captions = pd.read_excel('../data/captions_clean.xlsx')
df_captions.drop(columns =['Unnamed: 0', 'Long'], inplace = True)
df_captions
```

```python
df_merge2 = df_merge1.merge(df_captions, how = 'left', on = 'Short')
df_merge2
```

```python
# Check for duplicates in 'Column'
duplicates = df_merge2.duplicated('Long', keep=False)

# Display the duplicated values
print(df_merge2[duplicates])
```

```python
# Drop duplicates in 'Column1'
df_unique = df_merge2.drop_duplicates(subset='Long', keep='first')
df_unique
```
