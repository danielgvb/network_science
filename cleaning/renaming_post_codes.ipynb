```python
import os

def get_jpg_files(folder_path):
    jpg_files_dict = {}
    for root, dirs, files in os.walk(folder_path):
        for dir in dirs:
            dir_path = os.path.join(root, dir)
            jpg_files = [f for f in os.listdir(dir_path) if f.endswith('.jpg')]
            if jpg_files:
                jpg_files_dict[dir] = jpg_files
    return jpg_files_dict

# Example usage
folder_path = "../Avances Rafa/data/raw"
jpg_files_dict = get_jpg_files(folder_path)
```

```python
def match_list_to_dict_values(my_list, my_dict):
    matched_dict = {}
    for item in my_list:
        for key, value in my_dict.items():
            if item in value:
                matched_dict[item] = key
                break
    return matched_dict

# Example usage:
my_list = list(df_captions_rafa['Long'])
my_dict = jpg_files_dict 
matched_dict = match_list_to_dict_values(my_list, my_dict)
matched_dict
```
