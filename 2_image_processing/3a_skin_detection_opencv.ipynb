```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("sample/2020-12-21_18-09-40_UTC.jpg")
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h = hsv[:,:,0]
s = hsv[:,:,1]
v = hsv[:,:,2]
hsv_split = np.concatenate((h,s,v), axis = 1)



ret, min_sat = cv2.threshold(s,40,255,cv2.THRESH_BINARY)
ret, max_hue = cv2.threshold(h,15,255,cv2.THRESH_BINARY_INV)
final = cv2.bitwise_and(min_sat,max_hue)

'''plt.imshow(cv2.cvtColor(min_sat, cv2.COLOR_BGR2RGB))
plt.title('Sat Filter')
plt.show()

plt.imshow(cv2.cvtColor(max_hue, cv2.COLOR_BGR2RGB))
plt.title('Hue Filter')
plt.show()
'''
plt.imshow(cv2.cvtColor(final, cv2.COLOR_BGR2RGB))
plt.title('Final Filter')
plt.show()
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.show()
```

```python
# Count the number of white pixels (skin pixels)
white_pixels = np.sum(final == 255)

# Calculate the total number of pixels
total_pixels = final.size

# Calculate the percentage of white pixels
percentage_skin = (white_pixels / total_pixels) * 100

print(f"Percentage of skin in the image: {percentage_skin:.2f}%")
```
